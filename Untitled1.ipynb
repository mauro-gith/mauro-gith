```python
# prompt: crea grafico a barre che metta in relazione popolazione posizione geografica ricchezza e alfabetizzazione

import matplotlib.pyplot as plt
import numpy as np

# Sample data (replace with your actual data)
countries = ['italy', 'france', 'germany', 'spain', 'england']
population = [1000, 1200, 1600, 900, 1000]
latitude = [41, 48, 40, 52, 51]  # Example latitude values
wealth = [25000, 48000, 51000, 40000, 46000]  # Example wealth values (e.g., GDP per capita)
literacy_rate = [98 , 99 , 99 ,  97 , 99 ,]  # Example literacy rates

x = np.arange(len(countries))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(x - width, population, width, label='Population')
rects2 = ax.bar(x, latitude, width, label='Latitude')
rects3 = ax.bar(x + width, wealth, width, label='Wealth')

# Add literacy rate as text annotations above the bars
for i, rate in enumerate(literacy_rate):
  ax.text(x[i], wealth[i] +1000 , f'{rate:.2f}', ha='center', va='bottom') #Adjust 1000 to fit your data

ax.set_ylabel('Values')
ax.set_title('Population, Location, Wealth, and Literacy Rate by Country')
ax.set_xticks(x)
ax.set_xticklabels(countries, rotation=45, ha='right')
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)

fig.tight_layout()

plt.show()
```
