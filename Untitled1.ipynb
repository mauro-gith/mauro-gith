```python
# prompt: crea grafico a torta per mettere in relazione le vendite di a b c prodotti negli anni da 2022 a 2024 e l'aumento dei costi energetici

import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
years = [2022, 2023, 2024]
products = ['A', 'B', 'C']
sales = {
    2022: {'A': 100, 'B': 150, 'C': 120},
    2023: {'A': 120, 'B': 180, 'C': 130},
    2024: {'A': 150, 'B': 200, 'C': 160}
}
energy_costs = [10, 15, 20]  # Example energy cost increase percentages

# Create subplots for each year
fig, axes = plt.subplots(1, len(years), figsize=(15, 5))

for i, year in enumerate(years):
    labels = [f"{product} ({sales[year][product]})" for product in products]
    sizes = list(sales[year].values())

    # Explode the largest slice for emphasis
    explode = [0.1 if sizes[j] == max(sizes) else 0 for j in range(len(sizes))]

    axes[i].pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90)
    axes[i].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    axes[i].set_title(f"Sales in {year} (Energy Cost Increase: {energy_costs[i]}%)")

plt.tight_layout()
plt.show()
```
